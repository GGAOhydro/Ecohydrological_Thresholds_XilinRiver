```python
import os

# Define paths and output folder
template = 'your_path_here'
points = 'your_path_here'
out_folder = 'your_path_here'

# Ensure output directory exists
if not os.path.exists(out_folder):
    os.makedirs(out_folder)

# Initialize result file
results_file_path = os.path.join(out_folder, 'aa_0.txt').replace('\\', '/')
with open(results_file_path, 'w+') as results_file:

    # Set number of points to process
    point_num = 466
    start_num = 1

    # Write total points to results file
    results_file.write(str(point_num) + '\n')

    # Read the entire template file, including blank lines
    with open(template, 'r') as file:
        temp = [x.rstrip() for x in file.readlines()]  # Preserve line endings

    # Read data points file
    with open(points, 'r') as file:
        data = [x.strip().split(',') for x in file.readlines()]

    # Process each point
    for i in range(start_num, start_num + point_num):
        out_file_path = os.path.join(out_folder, f'point_{i}.dnd').replace('\\', '/')
        with open(out_file_path, 'w+') as out_file:
            for j, line in enumerate(temp):
                template_parts = line.split()  # Split the current line for processing
                if len(template_parts) > 1 and len(data) > j and len(data[j]) > i:
                    # Check if we have a valid line to update and if data exists for this line
                    original = template_parts[1]
                    template_parts[1] = data[j][i].strip()
                    print(f'Update {template_parts[0]} for point {i} from {original} to {template_parts[1]}')
                    out_file.write(' '.join(template_parts) + '\n')
                else:
                    # Write the original line, whether it's blank or not modifiable
                    out_file.write(line + '\n')
                    if line.strip():
                        print(f'Skipped {template_parts[0] if len(template_parts) > 0 else "Blank line"} for point {i}')

            results_file.write(out_file_path + '\n')
            print(out_file_path)
```
